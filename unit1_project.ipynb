```python
from transformers import pipeline, set_seed
```

```python
generator = pipeline(
    "text-generation",
    model="gpt2-medium",
    device=0  # Uses GPU if available
)

set_seed(42)
```

```python
def generate_poetry(theme, max_length=120):
    prompt = f"""
Write a short poem with rhyming lines about the theme "{theme}".
The poem should have vivid imagery and emotional depth.

Poem:
"""

    output = generator(
        prompt,
        max_length=max_length,
        num_return_sequences=1,
        temperature=0.9,        # Creativity
        top_p=0.95,             # Nucleus sampling
        repetition_penalty=1.2, # Avoid repetition
        do_sample=True
    )

    poem = output[0]["generated_text"]
    return poem
```

```python
print(generate_poetry("Love"))
```

```python
print(generate_poetry("Space"))
```

```python
print(generate_poetry("Coding"))
```
