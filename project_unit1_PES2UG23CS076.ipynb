```python
from transformers import pipeline
```

```python
study_text = """
Generative AI is a branch of artificial intelligence that focuses on creating new content
such as text, images, audio, and video. Large Language Models (LLMs) are trained on vast
amounts of data using transformer architectures. These models can perform tasks such as
text generation, summarization, question answering, and translation.
"""
```

```python
summarizer = pipeline(
    "summarization",
    model="sshleifer/distilbart-cnn-12-6"
)

summary = summarizer(
    study_text,
    max_length=60,
    min_length=30,
    do_sample=False
)

print("SUMMARY:\n", summary[0]['summary_text'])
```

```python
qa_pipeline = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad"
)
```

```python
## Example
```

```python
question1 = "What is Generative AI?"
result1 = qa_pipeline(question=question1, context=study_text)
print("Q:", question1)
print("A:", result1['answer'])
```

```python
question2 = "What architecture do LLMs use?"
result2 = qa_pipeline(question=question2, context=study_text)
print("\nQ:", question2)
print("A:", result2['answer'])
```
