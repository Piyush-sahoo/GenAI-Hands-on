```python
from transformers import pipeline, set_seed
```

```python
generator = pipeline(
    "text-generation",
    model="gpt2"
)

set_seed(42)
```

```python
def generate_titles(niche):
    prompt = f"List of viral YouTube video titles about {niche}:\n1."

    output = generator(
        prompt,
        max_length=80,
        num_return_sequences=1,
        temperature=0.9,
        top_p=0.95
    )

    text = output[0]["generated_text"]
    lines = text.split("\n")

    titles = []
    for line in lines:
        if line.strip() and len(titles) < 5:
            titles.append(line.strip())

    return titles
```

```python
niche = "gaming"
titles = generate_titles(niche)

for t in titles:
    print(t)
```
