## Project Objective
To develop an IT support bot that provides solutions for common technical issues using text generation.

```python
from transformers import pipeline

support_bot = pipeline(
    "text2text-generation",
    model="google/flan-t5-small"
)
```

## User Query

```python
user_query = "My computer is running very slow"
prompt = f"""
You are an IT support assistant.
Give a short and simple solution for the following problem.

Problem:
{user_query}
"""
```

## Bot Response

```python
response = support_bot(prompt, max_new_tokens=50)

print("IT Support Bot Response:\n")
print(response[0]['generated_text'])
```

## Observation
The IT support bot generated a relevant response based on the given technical issue.

## Final Understanding
This project shows how text generation can be used to handle basic IT support queries.
