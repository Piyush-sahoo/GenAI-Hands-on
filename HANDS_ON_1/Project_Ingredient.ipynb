```python
from transformers import pipeline
import re
```

```python
ner_pipeline = pipeline(
    "ner",
    model="dslim/bert-base-NER",
    aggregation_strategy="simple"
)
```

```python
blog_text = """
Last Sunday, I decided to cook a comforting dinner for my family. 
I went to the local market and bought fresh tomatoes, onions, garlic, and basil. 
On my way home, I also picked up some chicken, olive oil, butter, and cream. 
For dessert, I planned a simple dish using chocolate, sugar, and milk. 
The aroma of garlic and butter filled the kitchen as I started cooking.
"""
```

```python
ner_results = ner_pipeline(blog_text)

ner_results
```

```python
raw_entities = [ent["word"] for ent in ner_results]

print("Raw extracted entities:")
raw_entities
```

```python
cleaned_entities = set()

for word in raw_entities:
    word = word.lower().strip()
    word = re.sub(r"[^a-z\s]", "", word)
    if len(word) > 2:
        cleaned_entities.add(word)

cleaned_entities
```

```python
food_keywords = {
    "tomatoes", "onions", "garlic", "basil", "chicken",
    "olive oil", "butter", "cream", "chocolate",
    "sugar", "milk"
}

final_foods = []

for word in cleaned_entities:
    if word in food_keywords:
        final_foods.append(word)

final_foods
```

```python
# 1. Run NER on the blog text
ner_results = ner_pipeline(blog_text)

# 2. Extract and clean entities
raw_entities = [ent["word"] for ent in ner_results]
cleaned_entities = set()
for word in raw_entities:
    word = word.lower().strip()
    word = re.sub(r"[^a-z\s]", "", word)
    if len(word) > 2:
        cleaned_entities.add(word)

# 3. Define a food keyword set (expand as needed)
food_keywords = {
    "tomatoes", "onions", "garlic", "basil", "chicken",
    "olive oil", "butter", "cream", "chocolate",
    "sugar", "milk"
}

# 4. Extract foods present in the blog story
final_foods = [word for word in cleaned_entities if word in food_keywords]

# 5. Print results
print("✅ Extracted Ingredients from Blog Story:\n")
for item in sorted(final_foods):
    print("-", item)
```
