```python
from transformers import pipeline
```

```python
models = {
    "BERT": "bert-base-uncased",
    "RoBERTa": "roberta-base",
    "BART": "facebook/bart-base"
}
```

```python
prompt = "The future of Artificial Intelligence is"

for name, model in models.items():
    print("\n====================")
    print("Model:", name)
    try:
        generator = pipeline("text-generation", model=model)
        result = generator(prompt, max_length=30)
        print(result)
    except Exception as e:
        print("FAILED ❌")
        print(e)
```

```python
from transformers import pipeline

models = {
    "BERT": "bert-base-uncased",
    "RoBERTa": "roberta-base",
    "BART": "facebook/bart-base"
}

sentences = {
    "BERT": "The goal of Generative AI is to [MASK] new content.",
    "RoBERTa": "The goal of Generative AI is to <mask> new content.",
    "BART": "The goal of Generative AI is to <mask> new content."
}

for name, model in models.items():
    print("\n====================")
    print("Model:", name)
    fill = pipeline("fill-mask", model=model)
    results = fill(sentences[name])

    for r in results[:3]:
        print(r["sequence"], "| score:", r["score"])
```

## Observation Table

| Task | Model | Success / Failure | Observation | Architectural Reason |
|------|------|------------------|-------------|----------------------|
| Text Generation | BERT | Failure | Generated repetitive or meaningless symbols | Encoder-only model, not trained for text generation |
|  | RoBERTa | Failure | Could not generate meaningful continuation | Encoder-only architecture |
|  | BART | Partial | Generated incoherent or noisy text | Encoder–Decoder model, but not trained as causal LM |
| Fill Mask | BERT | Success | Predicted correct words like create, generate | Trained using Masked Language Modeling (MLM) |
|  | RoBERTa | Success | High-confidence predictions for missing word | Optimized MLM encoder |
|  | BART | Partial | Lower confidence predictions | Seq2Seq model, MLM is not primary objective |
| Question Answering | BERT | Partial | Extracted incomplete answer | Base model, not fine-tuned for QA |
|  | RoBERTa | Partial | Unstable or partial answer | No QA-specific fine-tuning |
|  | BART | Partial | Inconsistent answer | Requires task-specific fine-tuning |
