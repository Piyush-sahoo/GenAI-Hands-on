Cell 2: Set API Key

```python
from google import genai

client = genai.Client(api_key="<your API>")
```

Cell 3: Prompt Function

```python
def build_prompt(ingredients):
    return f"""
You are a professional home chef.

Using ONLY the following three ingredients: {ingredients},
write a single cooking instruction paragraph (80â€“120 words).

Rules:
- Do not add extra ingredients
- No bullet points or numbering
- Simple English
- Output only one paragraph
"""
```

Cell 4: Recipe Generator

```python
def generate_recipe(ingredients):
    if len(ingredients.split(",")) != 3:
        return "Error: Please enter exactly 3 ingredients."

    response = client.models.generate_content(
        model="models/gemini-flash-latest",
        contents=build_prompt(ingredients)
    )

    return response.text
```

Cell 5: Test the Project

```python
ingredients = "Eggs, Tomato, Cheese"
print(generate_recipe(ingredients))
```
