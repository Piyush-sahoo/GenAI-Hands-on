```python
import requests
from transformers import pipeline
from newspaper import Article, Config
import nltk

# Essential for parsing
nltk.download('punkt_tab', quiet=True)

def tldr_news_summarizer(input_data, is_url=True):
    user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
    config = Config()
    config.browser_user_agent = user_agent

    summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")

    if is_url:
        article = Article(input_data, config=config)
        article.download()
        article.parse()

        # VALIDATION: Check if we actually got the article or just a 404/Error page
        if len(article.text) < 100:
            return "Error: Article content too short. The link might be dead or blocked."

        article_text = article.text
        print(f"Summarizing: {article.title}")
    else:
        article_text = input_data

    # Dynamic length adjustment to avoid the warning you saw
    input_length = len(article_text.split())
    max_len = min(100, int(input_length * 0.5))

    summary = summarizer(article_text, max_length=max_len, min_length=20, truncation=True)
    return summary[0]['summary_text']

# Use a valid, recent URL
valid_url = "https://www.bbc.com/news/articles/c4g44r22j5jo"
print(tldr_news_summarizer(valid_url))
```

**My Learnings**  
First, the code visits the website you give it. Since big sites like the BBC often block basic scripts, we gave the code a "User-Agent" which is like a fake ID that makes the website think a real person is visiting from a normal browser. After it successfully reads the page, it passes the text to a model called DistilBART. This is a fast, lightweight AI that was specifically trained to condense long articles. It picks out the most important facts and rewrites them into a short, three-sentence summary.
