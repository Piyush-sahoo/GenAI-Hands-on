```python
from transformers import pipeline
bert_name = "bert-base-uncased"
roberta_name = "roberta-base"
bart_name = "facebook/bart-base"
```

 **Experiment 1: Text Generation**

```python
prompt = "The future of Artificial Intelligence is"
for model_name in [bert_name, roberta_name, bart_name]:
    try:
        generator = pipeline('text-generation', model=model_name)
        result = generator(prompt, max_length=20, num_return_sequences=1)
        print(f"Model: {model_name}\nOutput: {result[0]['generated_text']}\n")
    except Exception as e:
        print(f"Model: {model_name} failed. Error: {e}\n")
```

**Experiment 2: Masked Language Modeling**

```python
prompts = {
    "bert-base-uncased": "The goal of Generative AI is to [MASK] new content.",
    "roberta-base": "The goal of Generative AI is to <mask> new content.",
    "facebook/bart-base": "The goal of Generative AI is to <mask> new content."
}

for model_name in [bert_name, roberta_name, bart_name]:
    try:
        unmasker = pipeline('fill-mask', model=model_name)
        result = unmasker(prompts[model_name])
        print(f"Model: {model_name}\nTop Prediction: {result[0]['token_str']}\n")
    except Exception as e:
        print(f"Model: {model_name} failed. Error: {e}\n")
```

**Experiment 3: Question Answering (QA)**

```python
qa_context = "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
qa_question = "What are the risks?"

for model_name in [bert_name, roberta_name, bart_name]:
    try:
        qa_model = pipeline('question-answering', model=model_name)
        result = qa_model(question=qa_question, context=qa_context)
        print(f"Model: {model_name}\nAnswer: {result['answer']}\nScore: {result['score']}\n")
    except Exception as e:
        print(f"Model: {model_name} failed. Error: {e}\n")
```

Architecture Matters: Encoders (BERT/RoBERTa) excel at understanding context (Fill-Mask) but are fundamentally unable to generate new text because they process sentences all at once rather than one word at a time.

Hybrid Advantages: BART (Encoder-Decoder) performed better across both generation and QA because its "decoder" part is designed for sequence-to-sequence transformation, unlike raw encoders.

The "Base" Model Limit: All models struggled with Question Answering because "Base" models are like raw brains—they know language but haven't been taught the "rules" of the QA task via fine-tuning.

Generative Coherence: The "gibberish" in BART's generation shows that even with a decoder, a model needs a large training scale (like GPT-4) or specific instruction-tuning to stay human-coherent.
