```python
from transformers import pipeline

# Load BART model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")


# Split long legal text into chunks
def chunk_text(text, chunk_size=350):
    words = text.split()
    for i in range(0, len(words), chunk_size):
        yield " ".join(words[i:i+chunk_size])


# Simplification function
def simplify_legal_text(text):
    final_output = ""

    for chunk in chunk_text(text):
        # Instruction forces explanation, not short summary
        prompt = (
            "Rewrite the following legal terms and conditions in clear, "
            "simple, everyday language as a full explanatory paragraph: "
            + chunk
        )

        summary = summarizer(
            prompt,
            max_length=150,   # longer paragraph
            min_length=60,    # prevents short answer
            do_sample=False
        )

        final_output += summary[0]['summary_text'] + " "

    return final_output.strip()


# ---------------- MAIN ---------------- #

print("‚öñÔ∏è Legal Jargon Simplifier (Paragraph Mode)")
print("Paste Terms & Conditions text (multi-line allowed). Press ENTER twice to finish:\n")

# Multi-line input
lines = []
while True:
    line = input()
    if line == "":
        break
    lines.append(line)

user_input = " ".join(lines)

# Generate simplified paragraph
result = simplify_legal_text(user_input)

print("\nüü¢ Simplified Explanation (Layman Paragraph):\n")
print(result)
```
