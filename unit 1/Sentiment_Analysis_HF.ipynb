```python
from transformers import pipeline
```

```python
sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)
```

```python
text = "The product quality is amazing and I love it!"
result = sentiment_analyzer(text)
result
```

```python
user_input = input("Enter customer feedback: ")
analysis = sentiment_analyzer(user_input)

label = analysis[0]['label']
score = analysis[0]['score']

print("\nSentiment:", label)
print("Confidence Score:", round(score, 2))
```

```python
user_input = input("Enter customer feedback: ")
analysis = sentiment_analyzer(user_input)

label = analysis[0]['label']
score = analysis[0]['score']

print("\nSentiment:", label)
print("Confidence Score:", round(score, 2))
```

```python
reviews = [
    "The product is terrible and waste of money",
    "Customer support was very helpful",
    "Not satisfied with the quality",
    "Excellent experience, will buy again"
]

results = sentiment_analyzer(reviews)

for review, result in zip(reviews, results):
    print(f"\nReview: {review}")
    print("Sentiment:", result['label'])
    print("Score:", round(result['score'], 2))
```

## Customer Feedback Sentiment Analyzer

This project uses a Hugging Face pretrained transformer model to classify
customer feedback as Positive or Negative.

### Model Used:
distilbert-base-uncased-finetuned-sst-2-english

### Technique:
 Sentiment Analysis using NLP
 Transformer-based text classification

### Applications:

This project demonstrates how pretrained transformer models from Hugging Face
can be used to efficiently analyze customer sentiment without training a model
