#**The AI Storyteller**

The goal is  to create a script that takes a classic story-starting prompt and generates a short, creative paragraph.

```python
from transformers import pipeline, set_seed
story_generator = pipeline('text-generation', model='gpt2')
set_seed(42)
prompt = "Once upon a time in a faraway land, a small robot discovered"

story = story_generator(
    prompt,
    max_length=100,
    num_return_sequences=1,
    do_sample=True,
    top_k=50,
    temperature=0.9
)

print(story[0]['generated_text'])
```

So we initialised a text generation pipeline using the GPT-2 model to create the creative content part based on the starting prompt given.By setting do_sample=True, the model avoids repetitive loops and chooses words based on probability. This change makes the narrative sound more natural and varied. In the end, it shows that the decoder-only architecture can 'dream up' and finish a sequence of text, serving as an automated storyteller.
