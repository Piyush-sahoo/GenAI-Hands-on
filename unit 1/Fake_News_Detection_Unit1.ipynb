```python
from transformers import pipeline

fake_news_classifier = pipeline(
    "text-classification",
    model="mrm8488/bert-tiny-finetuned-fake-news-detection"
)
```

```python
headline = "Breaking: Scientists discover miracle cure overnight!"

result = fake_news_classifier(headline)
print("Prediction:", result)
```

```python
sentiment_analyzer = pipeline("sentiment-analysis")
sentiment_result = sentiment_analyzer(headline)

print("Sentiment:", sentiment_result)
```

```python
ner_tagger = pipeline("ner", grouped_entities=True)
ner_result = ner_tagger(headline)

print("Named Entities:", ner_result)
```

```python
def analyze_headline(headline):
    print("\nHeadline:", headline)

    classification = fake_news_classifier(headline)
    sentiment = sentiment_analyzer(headline)
    entities = ner_tagger(headline)

    print("\nFake News Prediction:")
    print(classification)

    print("\nSentiment Analysis:")
    print(sentiment)

    print("\nNamed Entity Recognition:")
    print(entities)

# Test
analyze_headline("Breaking: Scientists discover miracle cure overnight!")
```
