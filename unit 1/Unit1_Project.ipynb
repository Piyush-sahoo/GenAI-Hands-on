```python
from transformers import pipeline

print("Loading BERT model for grammar correction...")
mask_filler = pipeline("fill-mask", model="bert-base-uncased")
print("Model loaded successfully!\n")
```

```python
# Grammar Correction Function
def correct_grammar(sentence, mask_position):
    """
    Corrects grammar by masking a specific word position

    Args:
        sentence: Original sentence with error
        mask_position: Index of word to correct (0-based)
    """
    words = sentence.split()
    words[mask_position] = "[MASK]"
    masked_sentence = " ".join(words)

    print(f"Original:  {sentence}")
    print(f"Masked:    {masked_sentence}")
    print("\nTop 5 predictions:")
    print("-" * 40)

    predictions = mask_filler(masked_sentence)

    for i, pred in enumerate(predictions[:5], 1):
        print(f"{i}. '{pred['token_str']}' (confidence: {pred['score']:.3f})")

    # Get best prediction
    best_word = predictions[0]["token_str"]
    corrected = masked_sentence.replace("[MASK]", best_word)

    print("-" * 40)
    print(f"Corrected: {corrected}\n")

    return corrected
```

```python
# Test Cases
# Test 1: Your example
print("=" * 50)
print("TEST 1: Verb form error")
print("=" * 50)
correct_grammar("I am go to school", mask_position=2)  # masks "go"

print("\n" + "=" * 50)
print("TEST 2: Wrong preposition")
print("=" * 50)
correct_grammar("I am good in math", mask_position=3)  # masks "in"

print("\n" + "=" * 50)
print("TEST 3: Article error")
print("=" * 50)
correct_grammar("I saw a elephant", mask_position=2)  # masks "a"
```
