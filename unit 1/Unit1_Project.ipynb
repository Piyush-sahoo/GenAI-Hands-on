```python
from transformers import pipeline

# Initialize the sentiment analysis pipeline with the smallest standard model
# This model is optimized for speed and binary classification (Positive/Negative)
analyzer = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python
# Sample reviews representing 'happy' and 'angry' feedback
reviews = [
    "The product is amazing! It changed my life.",
    "Absolute waste of money. It broke on the first day.",
    "I'm quite happy with the quality, though shipping was a bit slow.",
    "Worst customer service ever. They ignored my emails for a week!",
    "Simple, elegant, and does exactly what it says on the box."
]

# Run the analysis on the entire list
# The pipeline handles the batching automatically
results = analyzer(reviews)

# Categorize and count results to see if the overall sentiment is "happy" or "angry"
happy_count = 0
angry_count = 0

print("--- Analysis Results ---")
for text, result in zip(reviews, results):
    label = result['label']
    score = result['score']

    # Map model labels to user categories
    sentiment_category = "HAPPY" if label == "POSITIVE" else "ANGRY"

    if sentiment_category == "HAPPY":
        happy_count += 1
    else:
        angry_count += 1

    print(f"Review: \"{text[:50]}...\"")
    print(f"Outcome: {sentiment_category} (Confidence: {score:.4f})\n")

print(f"Total Summary: {happy_count} Happy | {angry_count} Angry")
```
