FIRST LETS IMPORT ALL LIBRARIES 

```python
from transformers import pipeline,set_seed
```

LETS LOAD OUR MODEL (GPT2)

since we are focused on text generation we are using only a decoder model that is gpt2

```python
david_goggins = pipeline("text-generation",model="gpt2")
```

SETING UP AN ARRAY OF WORDS

defining all themes on which the user can ask motivational quotes on 


```python
themes = [
    "Success",
    "Grind",
    "Peace",
    "Discipline",
    "Failure",
    "Consistency",
    "Focus",
    "Confidence",
    "Growth",
    "Patience"
]
```

THE MAIN QUOTE GENERATION FUCNTION

```python
def asking_goggins(keyword, num_quotes=3):
    prompt = f"Give ONE short motivational quote about {keyword}.\nQuote:"
    
    outputs = david_goggins(
        prompt,
        max_new_tokens=30,
        temperature=0.8,
        do_sample=True,
        num_return_sequences=num_quotes
    )
    
    quotes = []
    for i, out in enumerate(outputs, 1):
        quote = out["generated_text"].replace(prompt, "").strip()
        quotes.append(f"{i}. {quote}")
    
    return quotes
```

ASKING USER INPUT AND RUNNING THE FUCNTION 

```python
print("Available themes:")
print(", ".join(themes))

user_keyword = input("\nEnter a keyword for motivation: ").strip().capitalize()

if user_keyword not in themes:
    print("Keyword not in predefined themes, generating anyway...\n")

quotes = asking_goggins(user_keyword, num_quotes=3)

print(f"\nMotivational Quotes on '{user_keyword}':\n")

for q in quotes:
    print(q)
```
