```python
from transformers import pipeline
```

```python
text_generator = pipeline(
    task="text-generation",
    model="gpt2"
)
```

```python
niche = input("Enter YouTube niche (e.g., Tech Reviews, Fitness): ")
```

```python

prompt = f"List of viral YouTube video titles about {niche}:\n"
```

```python
generated_output = text_generator(
    prompt,
    max_length=100,
    do_sample=True,
    temperature=0.9,
    top_k=50
)
```

```python
print("\nGenerated YouTube Video Titles:\n")
print(generated_output[0]["generated_text"])
```
