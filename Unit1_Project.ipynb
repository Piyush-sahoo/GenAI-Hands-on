```python
from transformers import pipeline

# Load model (you can swap later for better models)
generator = pipeline("text-generation", model="gpt2")
```

```python
def generate_captions(description, num=3):
    prompt = (
        "Example:\n"
        "Photo: Mountain sunrise\n"
        "Captions:\n"
        "1. Early light and quiet thoughts\n"
        "2. Nature’s alarm clock\n"
        "3. The sky decided to glow today\n\n"
        f"Photo: {description}\n"
        "Captions:\n"
        "1. "
    )

    result = generator(
        prompt,
        max_new_tokens=60,
        temperature=0.9,
        top_p=0.9,
        do_sample=True,
        pad_token_id=50256
    )[0]["generated_text"]

    text = result[len(prompt):].strip()
    lines = [l.strip() for l in text.split("\n") if l.strip()]

    captions = []
    for line in lines:
        if line[0].isdigit() and "." in line:
            line = line.split(".",1)[1].strip()
        captions.append(line)
        if len(captions)==num:
            break

    return captions
```

```python
desc = "Sunset at the beach"
captions = generate_captions(desc)

print("Generated Captions:\n")
for i, c in enumerate(captions, 1):
    print(f"{i}. {c}")
```
