```python
from transformers import pipeline

# Load Question Answering pipeline
qa_pipeline = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad"
)

# Context document (can be HR policy, IT manual, college rules, etc.)
context = """
Employees are entitled to 12 days of paid leave per year.
Sick leave can be taken for up to 5 days with a medical certificate.
Passwords must be changed every 90 days.
Work from home is allowed up to 2 days per week.
"""

print("Automated FAQ Bot")
print("Type 'exit' to quit.\n")

# Takes user questions in a loop
while True:
    question = input("Ask your question: ")

    if question.lower() == "exit":
        print("Exiting FAQ Bot.")
        break

    result = qa_pipeline(
        question=question,
        context=context
    )

    print(f"Answer: {result['answer']}")
    print(f"Confidence: {round(result['score'], 2)}\n")
```
