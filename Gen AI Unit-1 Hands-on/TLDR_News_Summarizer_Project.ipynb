```python
from transformers import pipeline
```

## LOAD SUMMARIZATION MODEL

```python
summarizer = pipeline(
    "summarization",
    model="sshleifer/distilbart-cnn-12-6"
)
```

## Input News Article


```python
article = """
Generative AI is rapidly transforming industries by enabling machines to generate human-like text, images, and code.
It is being widely adopted in healthcare, education, finance, and entertainment.
However, concerns related to data privacy, bias, and misinformation remain significant challenges.
Experts believe that responsible AI development and regulation are essential for long-term success.
"""
```

```python
summary = summarizer(
    article,
    max_length=50,
    min_length=10,
    do_sample=False
)

print("TL;DR Summary:\n")
print(summary[0]["summary_text"])
```

### Observation

The model successfully generated a short and meaningful summary from the given article.
This demonstrates how summarization models can condense long text while preserving important information.
